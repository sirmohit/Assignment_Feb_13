Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Ans: We have to use the Exception class while creating a Custom Exception because by extending the Exception class,
     our custom exception inherits all the properties and behaviors of the base Exception class. This means that we
     can utilize the built-in exception handling mechanisms provided by the language for the Exception class.
    
    Using the Exception class as the base for creating custom exceptions ensures consistency, compatibility, and 
    extensibility while leveraging the built-in exception handling mechanisms.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans:LookupError(StandardError) is used as a base class for exceptions raised when a sequence or dictionary type
    doesn't contain a given index or key.

```python
#Example
#KeyError : KeyError occurs in the program when we try to access or delete the dictonary key that does not exist.
dict = {1:"ram",2:"syam",3:"mohan"}
try:
    value=dict[4]
except KeyError as e:
    print(f"Error:{e}")
    
```

```python
#Example
#IndexError : IndexError occur when we try to access or delete the element which does not exist in the collection.
lst = [12,23,34,45,56]
try:
    val = lst[10]
    print(val)
except IndexError as e:
    print(e)
```

Q6. List down some best practices for exception handling in python.

Ans: Some best practise to handle the exception in python are:
     Always catch exceptions that you know how to handle and let the rest propagate up.
     Use the most specific exceptions possible in your except clauses.
     Avoid catching the base Exception class, as this can mask real issues in your code.
     Use the finally clause for cleanup actions that must always be completed.
